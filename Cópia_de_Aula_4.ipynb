## Laço infinito

```python
soma = 0

while True:
    numero = int(input("Digite um número inteiro (0 para sair): "))
    if numero == 0:
        break # Interrompe a loop infinito
    else:
      soma = soma + numero
      #soma += numero

print(f"A soma dos números inseridos é: {soma}")
```

Exemplo - for e else juntos (verificar se o numero 5 está ná lista)

```python
numeros = [1, 2, 3, 4, 6, 7, 8, 9, 10]

for numero in numeros:
    if numero == 5:
        print("Número 5 encontrado na lista!")
        break
else:
    print("Número 5 não está na lista.")
```

```python
import random

# Gera um número aleatório entre 1 e 10
numero_aleatorio = random.randint(1, 10)

acertou = False
while not acertou:
  n = int(input('Digite um numero entre 1 e 10: '))
  if numero_aleatorio > n:
    print(f'O número é maior que {n}')
  elif numero_aleatorio < n:
    print(f'O número é menor que {n}')
  else:
    print('Você acertou!')
    acertou = True
```

```python
import random

# Gera um número aleatório entre 1 e 50
numero_aleatorio = random.randint(1, 10000)

tentativas = 15
acertou = False

for i in range(tentativas):
  n = int(input(f'Tentativa {i+1} de {tentativas}. Digite um numero entre 1 e 10000: '))
  if numero_aleatorio > n:
    print(f'O número é maior que {n}')
  elif numero_aleatorio < n:
    print(f'O número é menor que {n}')
  else:
    print('Você acertou!')
    acertou = True
    break

if not acertou:
  print(f'Você não acertou. O número correto era {numero_aleatorio}')
```

```python
import random

nome_produto = input('Digite o nome do produto: ')
quantidade_inicial = int(input('Digite a quantidade inicial em estoque: '))

estoque = quantidade_inicial

while True:
  opcao = input('Deseja registrar uma venda (v), ver o estoque (e) ou sair (s)? ')

  if opcao.lower() == 'v':
    quantidade_venda = int(input('Digite a quantidade vendida: '))
    if quantidade_venda <= estoque:
      estoque -= quantidade_venda
      print('Venda registrada com sucesso!')
    else:
      print('Quantidade insuficiente em estoque.')
  elif opcao.lower() == 'e':
    print(f'Estoque atual de {nome_produto}: {estoque}')
  elif opcao.lower() == 's':
    break
  else:
    print('Opção inválida.')

print('Relatório final:')
print(f'Produto: {nome_produto}')
print(f'Quantidade restante em estoque: {estoque}')
```
